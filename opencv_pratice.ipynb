First, let's play with openCV and generate feature maps of John Petrucci

```python
from sklearn.datasets import load_sample_image
import numpy as np
import os
import cv2
import matplotlib.pyplot as plt
import 
```

```python
jp = cv2.imread("data/jp.jpg", cv2.IMREAD_GRAYSCALE) / 255
```

```python
f = plt.figure()
f.set_figwidth(8)
f.set_figheight(8)
plt.imshow(jp, cmap="gray")
```

Define receptive field dimensions, and stride sizes

```python
f_h = 10
f_w = 10

s_h = 10
s_w = 10
```

Find max number of strides in both directions

```python
N_h = int(jp.shape[0] / s_h)
N_w = int(jp.shape[1] / s_w)
```

Pixelate the image for fun

```python
jp_pix = np.zeros((N_h, N_w))

for i in range(N_h):
    for j in range(N_w):
        jp_pix[i, j] = np.mean(jp[s_h*i:i*s_h+f_h-1, s_w*j:j*s_w+f_w-1])
```

```python
f = plt.figure()
f.set_figwidth(8)
f.set_figheight(8)
plt.imshow(jp_pix, cmap="gray")
```
